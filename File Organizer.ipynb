1. choose a file
2. list down extensions
3. move to specific folders
 


```python
import os
import shutil

def organize_downloads(download_dir, destination_dir):
    os.chdir(download_dir)

    # Define file type categories
    audio = ('.aac', '.aiff', '.flac', '.m4a', '.mp3', '.ogg','.wav', '.wma', '.amr', '.ape', '.midi', '.au', '.ra', '.dsd', '.opus')
    video = ('.mp4', '.mkv', '.avi', '.mov', '.wmv', '.flv', '.webm','.m4v', '.3gp', '.ogv', '.mpeg', '.mpg', '.ts', '.vob', '.rm', '.rmvb')
    image = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff','.webp', '.svg', '.ico', '.jfif', '.heif', '.bat', '.raw', '.eps')
    archive = ('.zip', '.rar', '.tar.gz', '.tar.bz2')
    document =  ('.doc', '.docx', '.pdf', '.txt', '.ppt', '.pptx', '.xls', '.xlsx')

    # Functions to check file types
    def is_audio(file):
        return os.path.splitext(file)[1] in audio
        
    def is_video(file):
        return os.path.splitext(file)[1] in video

    def is_image(file):
        return os.path.splitext(file)[1] in image

    def is_archive(file):
        return os.path.splitext(file)[1] in archive

    def is_document(file):
        return os.path.splitext(file)[1] in document

    def is_screenshot(file):
        name, ext = os.path.splitext(file)
        return (ext in image) and "screenshot" in name.lower()

    # Create destination directories if they don't exist
    audio_destination = os.path.join(destination_dir, 'audio')
    video_destination = os.path.join(destination_dir, 'video')
    image_destination = os.path.join(destination_dir, 'image')
    archive_destination = os.path.join(destination_dir, 'archive')
    document_destination = os.path.join(destination_dir, 'document')
    screenshot_destination = os.path.join(destination_dir, 'image', 'screenshots')
    other_destination = os.path.join(destination_dir, 'other')

    os.makedirs(audio_destination, exist_ok=True)
    os.makedirs(video_destination, exist_ok=True)
    os.makedirs(image_destination, exist_ok=True)
    os.makedirs(archive_destination, exist_ok=True)
    os.makedirs(document_destination, exist_ok=True)
    os.makedirs(screenshot_destination, exist_ok=True)
    os.makedirs(other_destination, exist_ok=True)

    # Organize files
    for file in os.listdir():
        if is_audio(file):
            shutil.move(file, audio_destination)
        
        elif is_video(file):
            shutil.move(file, video_destination)
        
        elif is_image(file):
            if is_screenshot(file):
                shutil.move(file, screenshot_destination)
            else:
                shutil.move(file, image_destination)
        
        elif is_archive(file):
            shutil.move(file, archive_destination)
            
        elif is_document(file):
            shutil.move(file, document_destination)
        
        else:
            shutil.move(file, other_destination)

if __name__ == "__main__":
    # Example usage
    download_dir = input("Enter the path of your downloads directory: ")
    destination_dir = input("Enter the path of your destination directory: ")
    organize_downloads(download_dir, destination_dir)
```
